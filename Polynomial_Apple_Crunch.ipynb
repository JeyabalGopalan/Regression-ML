# Predicting Apple Sales Based on Discounts Percentage

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:/Users/jeyaj/Downloads/Overall Notes/Data/grocery_polynomial_data.csv")
print(df)
```

```python
df.info()
```

```python
# Ensure numeric values
df['Discount Percentage'] = pd.to_numeric(df['Discount Percentage'], errors='coerce')
df['Apple Sales'] = pd.to_numeric(df['Apple Sales'], errors='coerce')
df = df.dropna(subset=['Discount Percentage', 'Apple Sales'])
```

## Apple sales initially increase with discounts but later stabilize or decline due to customer suspicion of stale products.

```python
# Prepare features and target
X = df[['Discount Percentage']].values.astype(float)
y = df[['Apple Sales']].values.astype(float)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## The best polynomial degree is 10

```python
# Transform data to include polynomial terms 
poly = PolynomialFeatures(degree=10)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Fit the model
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Predict on test data
y_pred = model.predict(X_test_poly)
```

```python
# Generate quadratic predictions for visualization
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_line_poly = poly.transform(X_line)
y_line_pred = model.predict(X_line_poly)
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))

# Plot training data and predictions
plt.scatter(X_train, y_train, color="blue", label="Training Data")
plt.plot(X_line, y_line_pred, color="purple", label="Model Prediction", linewidth=2)

# Plot testing data and predictions
plt.scatter(X_test, y_test, color="red", label="Testing Data")
plt.scatter(X_test, y_pred, color="green", label="Test Predictions", marker="x", s=100)

# Add titles and labels
plt.title("Polynomial Regression: Training and Testing Visualization")
plt.xlabel("Discount Percentage")
plt.ylabel("Apple Sales")
plt.legend()
plt.grid()
plt.show()
```
